```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#reading the datasets
df1 = pd.read_csv('MSCI_Data.csv')
df2 = pd.read_csv('All_Equities_Dataset.csv')

```

```python
df1.head()
```

```python
df1 = df1.set_index('Date')
```

```python
df2 = df2.set_index('Dates')
```

```python
df2.head()
```

```python
merged_df = pd.merge(df1, df2, left_index = True, right_index = True, how='inner')
print(merged_df.head())
```

```python
#Calculating the Returns
returns_data = merged_df.pct_change()
```

```python
returns_data.head()
```

```python
#Beta Calculation
#Defining our targets
targets = ['MSCI USA', 'MSCI USA Value', 'MSCI USA Low Volatility', 'MSCI USA Low Size',
           'MSCI USA High Dividend Yield', 'MSCI USA Quality', 'MSCI USA Momentum']
```

```python
#Defining Equities
equities = [col for col in returns_data.columns if col not in targets ]
equities
```

```python
#Calculating the 5 year rolling variance
rolling_var = returns_data[targets].rolling(window = 60).var()
```

```python
beta_storage = {}

# Using for loop across all the equities and then calculating the rolling covariance and ultimately the betas with our factors
for equity in equities:
    equity_beta = pd.DataFrame(index=returns_data.index)

    for target in targets:
        # Calculate rolling covariance between the specific equity and target
        # This ensures we get a Series, not a DataFrame with multiple columns
        rolling_covariance = returns_data[equity].rolling(window=60).cov(returns_data[target])
        
        # Divide by the rolling variance of the target
        equity_beta[f'Beta_{target}'] = rolling_covariance / rolling_var[target]
        equity_beta.dropna(inplace = True)

    beta_storage[equity] = equity_beta

beta_storage
```

```python
beta_df = pd.concat(beta_storage, axis=1)
beta_df.columns = ['_'.join(col).strip() for col in beta_df.columns.values]
beta_df.head()
```
